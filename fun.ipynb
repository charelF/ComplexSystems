# Game of Life

```python
N = 50
W = np.zeros((N,N))
W[1:8, 1:38] = 1

def update(W, i,j):
    neighbours = W[max(i-1,0):min(i+2,N), max(j-1,0):min(j+2,N)].flatten()
    count = sum(neighbours) - W[i,j]
    # print(count)
    if count < 2: W[i,j] = 0
    elif count == 3: W[i,j] = 1
    elif count > 3: W[i,j] = 0

prev = np.ones_like(W)
for t in range(50):
    if np.allclose(prev, W): break
    prev = W.copy()
    for i in range(N):
        for j in range(N):
            update(W, i, j)
    plt.figure(figsize=(5,5))
    plt.title(f"Iteration {t}")
    plt.imshow(W, interpolation="none", cmap="Greys")   
    plt.show()
    clear_output(wait=True)
```

# Logistic Map

```python
f = lambda x, mu: mu*x*(1-x)
N = 40

for mu, x0 in [
    (2.8, 0.2),
    (0.8, 0.7),
    (4, 0.2),
]:
    x = x0
    X = np.linspace(0,1,N)
    FX = f(X, mu)
    Y = np.zeros(N)
    for i in range(N):
        x = f(x, mu)
        Y[i] = x

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

    ax1.plot(X, FX, color="black", linestyle="--", label="y = logisticmap(x)")
    ax1.plot(X, X, color="black", linestyle=":", label="y = x")
    # ax1.plot(Y, np.roll(Y, -1), ".-")
    ax1.grid(alpha=0.2)
    # ax1.legend()

    # Code to generate cobweb plot
    # Source: https://scipython.com/blog/cobweb-plots/
    # is a bit more complicated because we want to plot straight lines between points
    # so for each x and y point we have to specify another x and y just to get a straight line
    # generate 2 arrays with (N+1) rows and 2 columns
    px, py = np.zeros((2,N+1,2))
    px[0], py[0] = x0, 0
    for i in range(1, N, 2):
        px[i] = px[i-1]  # x is the previous x
        py[i] = f(px[i-1], mu)  # y is f(previous x)
        px[i+1] = py[i]  # next x is current y
        py[i+1] = py[i]  # next y is current y

    # Plot the path traced out by the iteration.
    ax1.plot(px, py, ".-", color="blue", alpha=0.5)

    ax1.set_xlabel("x")
    ax2.set_xlabel("n")
    ax2.set_ylabel("$x_n$")

    ax2.plot(Y, ".-", color="blue", alpha=0.5)
    ax2.grid(alpha=0.2)
    fig.suptitle(f"x={x0}, μ={mu}")
    plt.tight_layout()
    plt.show()
```

# Logistic Bifurcation Diagram

```python
N = 100
lastX = 30
M = 2000

mus = np.array([[x]*lastX for x in np.linspace(2.5,4,M)])
periods = np.empty((M, lastX))

for i in range(M):
    mu = mus[i,0]
    x = 0.1
    Y = np.empty(N)
    for j in range(N):
        x = f(x, mu)
        Y[j] = x
    uni = np.unique(np.round(Y[-lastX:],3))
    periods[i] = np.pad(
        uni, (0, lastX-uni.shape[0]), 
        "constant", constant_values=(np.nan)
    )

plt.figure(figsize=(10,5))
plt.scatter(mus, periods, s=1, alpha=0.2)
plt.ylabel("Number of periods")
plt.xlabel("$\mu$")
plt.show()
```

# Lyapunov Exponent

```python
M = 1000
N = 100
x0 = 0.2

df = lambda x, mu: mu - 2*mu*x

mus = np.linspace(2.8,4,M)
lambdas = np.empty((M))

for i in range(M):
    mu = mus[i]
    x = x0
    Y = np.empty(N)
    for j in range(N):
        x = f(x, mu)
        Y[j] = x
    λ = (1/N) * sum([math.log(abs(df(xi, mu))) for xi in Y])
    lambdas[i] = λ

plt.figure(figsize=(10,5))
plt.plot(mus, lambdas)
plt.grid(alpha=0.2)
plt.ylabel("Lyapunov Expoenent ($\lambda$)")
plt.xlabel("$\mu$")
plt.plot(mus, np.zeros_like(mus), "--", color="black", alpha=0.5)
plt.show()
```

# Power Laws

```python
exp = lambda x, c: c**x
pwl = lambda x, c: x**c

c = 0.5
X = np.linspace(1, 100, 100)
EXPX = exp(X, 0.95)
PWLX = pwl(X, -0.5)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))

ax1.set_title("linear-linear")

ax2.set_title("log-linear")
ax2.set_yscale("log")

ax3.set_title("linear-log")
ax3.set_xscale("log")

ax4.set_title("log-log")
ax4.set_yscale("log")
ax4.set_xscale("log")

for ax in (ax1, ax2, ax3, ax4):
    ax.plot(EXPX, label="exponential")
    ax.plot(PWLX, label="power law")
    ax.legend()
    ax.grid(linestyle="--", alpha=0.4)

plt.suptitle("Exponential decay vs PowerLaw decay")
plt.tight_layout()
plt.show()

print("Exponential is linear in log-linear plot, Power-law is linear in log-log plot")
print("Exponential decays faster (halves each iteration when c=0.5) then power law")
```

# Network Giant Component Phase transition

```python
f = lambda s, avgk: 1 - np.exp(-avgk*s)
N = 40

for avgk, s0 in [
    (0.5, 0.5),
    (1.0, 0.5),
    (1.5, 0.2),
    (1.5, 0.7),
]:
    s = s0
    S = np.linspace(0,1,N)
    FS = f(S, avgk)
    Y = np.zeros(N)
    for i in range(N):
        s = f(s, avgk)
        Y[i] = s

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

    ax1.plot(S, FS, color="black", linestyle="--", label="y = f(x)")
    ax1.plot(S, S, color="black", linestyle=":", label="y = x")
    # ax1.plot(Y, np.roll(Y, -1), ".-")
    ax1.grid(alpha=0.2)
    # ax1.legend()

    # Code to generate cobweb plot
    # Source: https://scipython.com/blog/cobweb-plots/
    # is a bit more complicated because we want to plot straight lines between points
    # so for each x and y point we have to specify another x and y just to get a straight line
    # generate 2 arrays with (N+1) rows and 2 columns
    px, py = np.zeros((2,N+1,2))
    px[0], py[0] = s0, 0
    for i in range(1, N, 2):
        px[i] = px[i-1]  # x is the previous x
        py[i] = f(px[i-1], avgk)  # y is f(previous x)
        px[i+1] = py[i]  # next x is current y
        py[i+1] = py[i]  # next y is current y

    # Plot the path traced out by the iteration.
    ax1.plot(px, py, ".-", color="blue", alpha=0.5)

    ax1.set_xlabel("x")
    ax2.set_xlabel("n")
    ax2.set_ylabel("$x_n$")

    ax2.plot(Y, ".-", color="blue", alpha=0.5)
    ax2.grid(alpha=0.2)
    fig.suptitle(f"x={s0}, μ={avgk}")
    plt.tight_layout()
    plt.show()
```
