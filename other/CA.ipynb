```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, HTML
from matplotlib import animation, rc; rc('animation', html='jshtml')
import math
```

```python
def rulegen(num, r=1):
    states = [tuple([int(d) for d in f"{bin(i)[2:]:0>3}"]) for i in reversed(range(2**(2*r+1)))]
    rule = [int(d) for d in f"{bin(num)[2:]:0>{(2**(2*r+1))}}"]
    return {s:d for s,d in zip(states, rule)}
    return dic
```

```python
N = 150
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

toprow = np.round(np.random.random(size=N), 0)

for R, ax, i in zip([250, 108, 182, 110], axs.flatten(), range(1,5)):
    F = rulegen(R)
    M = np.zeros((N//2,N))
    M[0] = toprow
    for rowi in range(0,(N//2)-1):
        for coli in range(0,N):
            M[rowi+1, (coli+1)%N] = F[
                M[rowi,(coli)%N],
                M[rowi,(coli+1)%N],
                M[rowi,(coli+2)%N],
            ]

    ax.set_title(f"Class {i} (R{R})")
    ax.imshow(M, interpolation="none", cmap="Greys")

plt.tight_layout()
plt.show()
```

### Cellular automata

simplify components and interactions
(some are) computationally universal: ability to implement any finite algorithm and thus evaluate any computatabl function → all universal computers are formally equvlanet

Reasons to study CA:
- emergence of complex, systemic behavious out of simple, local dynamics
- As original models of fundamental physics
- Discrete Dynamical System Simulators
- As powerful computation engines

most CA have 5 characteristics: (book)
Discrete lattice of cells
Homogeneity: all cells equivalent
Discrete States (finite number possible)
Local interactions: cells only interact with local neighbourhood
Discrete dynamics: at discrete unit time, cell updates its current state according t transition rule taking into account states of cells in neighbourhood

1D CA:
cells have k possible values → ci(t) in {0,1,2,...,k}
