```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f2(T, M, Tc, a0, b0):
    # lhs = f(T,M) - f0(T)
    return a0 * (T-Tc) * abs(M)**2 + b0 * abs(M)**4

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

M = np.linspace(-2, 2, 100)

ax1.set_title("a=1, b=1")
ax1.plot(M, f2(1, M, .1, 1, 1), label="T > Tc")
ax1.plot(M, f2(1, M, 1, 1, 1), label="T = Tc")
ax1.plot(M, f2(1, M, 2, 1, 1), label="T < Tc")
ax1.legend()

ax2.set_title("a=5, b=1")
ax2.plot(M, f2(1, M, .1, 5, 1), label="T > Tc")
ax2.plot(M, f2(1, M, 1, 5, 1), label="T = Tc")
ax2.plot(M, f2(1, M, 2, 5, 1), label="T < Tc")
ax2.legend()

ax3.set_title("a=1, b=5")
ax3.plot(M, f2(1, M, .1, 1, 5), label="T > Tc")
ax3.plot(M, f2(1, M, 1, 1, 5), label="T = Tc")
ax3.plot(M, f2(1, M, 2, 1, 5), label="T < Tc")
ax3.legend()

plt.show()
```

```python
def f1(T, M, Tc, a0, b0, c0):
    # lhs = f(T,M) - f0(T)
    return a0 * (T-Tc) * abs(M)**2 + b0 * abs(M)**4 + c0 * abs(M)**6

fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(12, 16))

M = np.linspace(-2, 2, 100)

tmp = {
    0:(5,-1,1),
    1:(1,-5,1),
    2:(1,-1,2),
    3:(5,-5,1),
    4:(5,-5,1),
    5:(5,-1,2),
    6:(5,-5,2),
    7:(1,-1,1),
}

i = 0
for row in axs:
    for elm in row:
        elm.set_ylim(-10, 10)
        elm.grid(linestyle="--", alpha=.4)
        a,b,c = tmp[i]
        i+=1
        elm.set_title(f"a={a}, b={b}, c={c}")
        elm.plot(M, f1(1, M, .1, a,b,c), label="T > Tc")
        elm.plot(M, f1(1, M, 1, a,b,c), label="T = Tc")
        elm.plot(M, f1(1, M, 2, a,b,c), label="T < Tc")
        elm.legend()


plt.show()
```

```python
pl = lambda x,a,t: (a*x)**-t
ex = lambda x,a,t: t**(-a*x)

X = np.linspace(0,5,100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))


ax1.set_title("power law")
ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.plot(X,pl(X,3,2))
ax1.plot(X,pl(X,4,2))

ax2.set_title("exp")
# ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.plot(X,ex(X,3,2))
ax2.plot(X,ex(X,4,2))

ax3.set_title("linear")
ax3.plot(X,X*3)
ax3.plot(X,X*4)

ax4.text(s="have a nice day!", x=0.5, y=0.5)

plt.tight_layout()
plt.show()
```
